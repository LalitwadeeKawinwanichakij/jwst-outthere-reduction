```python
phot_cat=cos20_cat
eazy_phot_cat =Table()
eazy_phot_cat['id']=copy.deepcopy(phot_cat['ID'])
eazy_phot_cat['ra'] = copy.deepcopy(phot_cat['ALPHA_J2000'])
eazy_phot_cat['dec'] = copy.deepcopy(phot_cat['DELTA_J2000'])
for c in phot_cat.colnames:
    if 'FLUXERR' in c:
        band = c.replace('_FLUXERR','')
        print(band)
        eazy_phot_cat['f_%s' % band] = copy.deepcopy(phot_cat['%s_FLUX' % band]) # uJy
        eazy_phot_cat['e_%s' % band] = copy.deepcopy(phot_cat['%s_FLUXERR' % band]) # uJy
eazy_phot_cat['z_spec'] = copy.deepcopy(phot_cat['ez_z_phot'])
for col in eazy_phot_cat.colnames:
    if hasattr(eazy_phot_cat[col],'filled'):
        eazy_phot_cat[col]= eazy_phot_cat[col].filled(-99)
```

```python
eazy_phot_cat.write('cosmos2020_farmer_r1_v2.2_p3_4eazy.fits',overwrite=True)
```

```python
# ### Initialize **eazy.photoz** object for JADES
import pickle
args=  np.load('/Users/lkawinwanichakij/NISPureParallel/FIELDS/sex-07/Extractions/fit_args.npy', allow_pickle=True)[0]


from grizli.pipeline import auto_script,photoz
import eazy
params={}
params['CATALOG_FILE'] = '/Users/lkawinwanichakij/JWST/NIRISS/OutThere/outthere_sex_apcorphot/outthere_sex-00-32+astrodeep+cosmos2020_photcat_4eazy.fits'
params['Z_STEP'] = 0.002
params['Z_MAX'] = 12
params['MAIN_OUTPUT_FILE'] = 'sex-00-32+astrodeep+cosmos2020.eazypy'
params['PRIOR_FILTER'] = 205 # f160w
# Galactic extinction
#params['MW_EBV'] = {'aegis':0.0066, 'cosmos':0.0148, 'goodss':0.0069, 
#                    'uds':0.0195, 'goodsn':0.0103}[field]
params['MW_EBV'] = 0.0148
params['TEMPLATES_FILE'] = 'templates/fsps_full/tweak_fsps_QSF_12_v3.param'
params['PRIOR_FILE']='templates/prior_F160W_TAO.dat'
translate_file = '/Users/lkawinwanichakij/JWST/NIRISS/OutThere/outthere_sex_apcorphot/astrodeep+niriss+cosmos2020.translate'
jades_ez = eazy.photoz.PhotoZ(param_file=None, translate_file=translate_file, 
                        zeropoint_file=None, params=params, 
                        load_prior=True, load_products=False)
## First-pass redshift templates, similar to the eazy templates but 
## with separate emission lines
t0 = args['t0']
## Make a helper object for generating photometry in a format that grizli 
## understands.
phot_obj = photoz.EazyPhot(jades_ez , grizli_templates=t0, zgrid=jades_ez.zgrid)
with open('sex-00-32-astrodeep+niriss+cosmos2020_eazyphot.pkl', "wb") as file:  # Open the file in binary write mode
    pickle.dump(phot_obj, file)
```
